```python
import pandas as pd
from IPython.display import Markdown
from pandas import util
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
df = util.testing.makeTimeDataFrame()
```

```python
# | label: tbl-test-to-markdown-option
# | tbl-cap: Testing with `.to_markdown()`
# | output: true
from IPython.display import Markdown

display(Markdown(df.head().to_markdown()))
```

```python
markdown_string = (
    df.head()
    .style.set_table_styles([dict(selector="th", props=[("text-align", "center")])])
    .background_gradient(axis=0)
    .set_table_styles(
        {0: [{"selector": "td:hover", "props": [("font-size", "25px")]}]},
        axis=1,
        overwrite=False,
    )
    .bar(subset=["A", "B"], color="#d65f5f")
)
```

```python
# | label: tbl-test-to-markdown-pandas-styler
# | tbl-cap: Embedding a table with panda styler
# | output: true

html = markdown_string.to_html()
Markdown(html)
```

```python
# | label: tbl-test-to-markdown-option-quarto-disable
# | tbl-cap: Embedding a table using data-quarto-disable-processing
# | output: true

html = markdown_string.to_html().replace(
    "<table ", '<table data-quarto-disable-processing="true" '
)
Markdown(html)
```
